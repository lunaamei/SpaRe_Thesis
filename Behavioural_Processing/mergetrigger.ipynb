```python

# START: NOT NEEDED for Pilot 3!!

data, header = pyxdf.load_xdf('/Users/lunameidoering/Uni/THESIS/feelSpacedata/ERP/Pilot4/sub-P004_ses-S004_task-Default_run-001_eeg.xdf') # check if name works

# get the new time:

for s in data:
    # stream name
    s_name = s["info"]["name"][0]
    # check SR of the EEG data
    if "BackupAlignment" in s_name:
        time_eeg = s["time_stamps"][0]
    if "EyeTrackingPosDir" in s_name:
        et_start = s["time_stamps"][0]

# check if P100 will be delayed, then double check if we have additional delays/shifts to take care off
to_sub = et_start - time_eeg 
# correct unity with the different timestamps
time = for_eye["time"].tolist() + to_sub

# END: NOT NEEDED for Pilot 3!!

for_eye['time'] = time

# generate Trigger file


# create triggers as done in the code before

gaze = for_eye[for_eye["events"] == 2.0]
gaze = gaze[~gaze["long_events"].isnull()]

no_name = gaze[gaze["names"].isnull()]
gaze = gaze[~gaze["names"].isnull()]

# now we define the different categories for the different triggers
Global_Landmark_df = gaze[gaze["names"].isin(Global_Landmark)]
TaskBuilding_Public_df = gaze[gaze["names"].isin(TaskBuilding_Public)]
TaskBuilding_Residential_df = gaze[gaze["names"].isin(TaskBuilding_Residential)]
Other_Buildings_df = gaze[gaze["names"].isin(Other_Buildings)]
Background_df = gaze[~gaze["names"].isin(Global_Landmark+TaskBuilding_Public+TaskBuilding_Residential+
                                         Other_Buildings)] # '~' inverts the statement

# create the trigger df:
# first get all the starts in one column
lat = pd.concat([Global_Landmark_df["time"], TaskBuilding_Public_df["time"], TaskBuilding_Residential_df["time"],
                 Other_Buildings_df["time"], Background_df["time"], no_name["time"]])
lat = pd.DataFrame(lat)

# then add the type: 2 for background, 1 for NPCs and 0 for faces, 3 for no name
lat["type"] = (
    ["Global_Landmark"] * len(Global_Landmark_df["time"]) +
    ["TaskBuilding_Public"] * len(TaskBuilding_Public_df["time"]) +
    ["TaskBuilding_Residential"] * len(TaskBuilding_Residential_df["time"]) +
    ["Other_Buildings"] * len(Other_Buildings_df["time"]) +
    ["Background"] * len(Background_df["time"]) +
    ["no_name"] * len(no_name["time"])
)

#selected_columns = ['nodeNr','num_neighbours','closest_to_nodeCentroid','closest_to_nodeCentroid','node_1st_half', 'time_diff','time_diff_normalized']

lat["in_nodeRadius"] = Global_Landmark_df["in_nodeRadius"].tolist() + TaskBuilding_Public_df["in_nodeRadius"].tolist() + TaskBuilding_Residential_df["in_nodeRadius"].tolist() +  Other_Buildings_df["in_nodeRadius"].tolist() + Background_df["in_nodeRadius"].tolist() + no_name["in_nodeRadius"].tolist()
lat["closest_to_nodeCentroid"] = Global_Landmark_df["closest_to_nodeCentroid"].tolist() + TaskBuilding_Public_df["closest_to_nodeCentroid"].tolist() + TaskBuilding_Residential_df["closest_to_nodeCentroid"].tolist() +  Other_Buildings_df["closest_to_nodeCentroid"].tolist() + Background_df["closest_to_nodeCentroid"].tolist() + no_name["closest_to_nodeCentroid"].tolist()
lat["time_diff_normalized"] = Global_Landmark_df["time_diff_normalized"].tolist() + TaskBuilding_Public_df["time_diff_normalized"].tolist() + TaskBuilding_Residential_df["time_diff_normalized"].tolist() +  Other_Buildings_df["time_diff_normalized"].tolist() + Background_df["time_diff_normalized"].tolist() + no_name["time_diff_normalized"].tolist()
lat["node_1st_half"] = Global_Landmark_df["node_1st_half"].tolist() + TaskBuilding_Public_df["node_1st_half"].tolist() + TaskBuilding_Residential_df["node_1st_half"].tolist() +  Other_Buildings_df["node_1st_half"].tolist() + Background_df["node_1st_half"].tolist() + no_name["node_1st_half"].tolist()
lat["num_neighbours"] = Global_Landmark_df["num_neighbours"].tolist() + TaskBuilding_Public_df["num_neighbours"].tolist() + TaskBuilding_Residential_df["num_neighbours"].tolist() +  Other_Buildings_df["num_neighbours"].tolist() + Background_df["num_neighbours"].tolist() + no_name["num_neighbours"].tolist()

# sort it by start so that all the events are in correct order
lat = lat.sort_values(by=["time"])

# rename start to latency
lat = lat.rename(columns={"time": "latency"})


#lat.latency = lat.latency + (1.8942931946367025 - 0.08) # for pilot 3
lat.latency = lat.latency + .02 # for pilot 4

lat = lat.reset_index(drop=True)

lat.to_csv("/Users/lunameidoering/Uni/THESIS/feelSpacedata/GraphMaking/Output/TriggerFile_pilot4_GraphMaking.csv", index=False)
```
